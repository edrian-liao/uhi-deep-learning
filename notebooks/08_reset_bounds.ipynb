# Reset Bounds

- Durham and Raleigh enclosed in a single file

```python
import geopandas as gpd
import torch
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import rasterio as rio
from shapely.geometry import box
```

## Albuquerque

```python
city = 'Albuquerque'
gdf = gpd.read_file(f'../data/shapefiles/{city}/pm_trav.shp')
gdf.columns
```

```python
gdf = gdf.to_crs(epsg=3857) # in meters
bounds = gdf.total_bounds
bounds
```

```python
ominx, ominy, omaxx, omaxy = gdf.total_bounds
bbox_new = box(ominx, ominy, omaxx, omaxy)
bbox_original = box(ominx, ominy, omaxx, omaxy)

# Plot
fig, ax = plt.subplots(figsize=(8, 8))
gdf.plot(ax=ax, markersize=1, label='Points')
gpd.GeoSeries(bbox_new).plot(ax=ax, facecolor="none", edgecolor='red', linewidth=2, label='Convex Hull Box')
gpd.GeoSeries(bbox_original).plot(ax=ax, facecolor="none", edgecolor='blue', linewidth=2, linestyle='--', label='Original Bounds')

# Final plot touches
plt.legend()
plt.title(f"{city}", fontsize=32)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.tight_layout()
plt.show()
```
